# COUNTPLOT TO SHOW VARIOUS MODELS OF EACH CARS

# VIZUALZING THE DATA USING THE HEATMAP

```python
colors = ['violet', 'skyblue', 'green', 'yellow', 'orange', 'pink', 'brown']
corr = df.corr()

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)]=True
with sns.axes_style('white'):
    fig, ax = plt.subplots(figsize=(8,8))
    sns.heatmap(corr,  mask=mask, annot=True, cmap=colors, center=0, square=True)
```

# PLOTTING JOINT PLOT

```python
plt.figure(dpi = 100, figsize = (5,4))
print("Joint plot of price with Other Variables ==> \n")
for i in  df.columns:
    if i != 'tax' and i != 'model' and i != 'year' and i != 'transmission' and i != 'mileage' and i != 'fuelType' and i != 'tax' and i != 'mpg' and i != 'engineSize' and i != 'Make':
        print(f"Correlation between tax and {i} ==> ",df.corr().loc['tax'][i])
        sns.jointplot(x='tax',y=i,data=df,kind = 'reg',color = 'red')
        plt.show()
```

#### Joint Plots allow displaying the relationship between two variables as well as the 1D profile of both variables as shown below . We pass in the kind of graph we want to use to display the relationship between the two variables. In the first graph, we used a regression.
