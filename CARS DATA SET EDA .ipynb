#  In which year most cars got manufactured?

```python
#using countplot we are checking the total no.of cars getting manufactured each year
plt.subplots(figsize = (10,5))
sns.countplot(data = df, x = "year")
plt.title("cars getting manufactured each year", fontsize = 18)
plt.xlabel("year", fontsize = 14)
plt.ylabel("no. of cars manufactured",fontsize = 14)
```


from above graph we can conclude that most cars got manufactured in the year 2019
we can also say that the sale got increased rapidly in 2016 but got dropped in 2018 and got increased again in 2019.

```python
#plotting pie chart
df.transmission.value_counts().plot.pie()
plt.show
```

```python
cols = 3
rows = 3
num_cols = df.select_dtypes(exclude='object').columns
fig = plt.figure(figsize= (15,9))
for i, col in enumerate(num_cols):
    
    ax=fig.add_subplot( rows, cols, i+1)
    
    sns.boxplot(x=df[col], ax=ax)
    
fig.tight_layout()  
plt.show()
```

```python
# 1. HISTOGRAM - sns.histplot()
sns.histplot(x='mileage', data=df)
```

```python
sns.violinplot(x=df["year"])
```

```python
sns.pairplot(data=df , vars = ["price" , "engineSize" , "Make"])
plt.show()
```

```python
df.hist(figsize=(12, 9))
plt.show()
```

```python
plt.figure(figsize=(8, 6))
correlations = df.corr().round(2)
mask = np.triu(np.ones_like(correlations))
sns.heatmap(data=correlations, mask=mask, square=True, linewidths=0.5, annot=True)
plt.show()
```
